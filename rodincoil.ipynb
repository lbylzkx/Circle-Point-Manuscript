```python
import numpy as np
import matplotlib.pyplot as plt
from circle import *

fig, ax = plt.subplots(figsize=(12, 12))
ax.set_aspect('equal')
R = 1
r = 2*R
m = 48
color = ['r', 'orange', 'y', 'g', 'c', 'b', 'purple',
         'gold', 'grey', 'pink', 'brown', 'olive']*4
for n in range(0, m):
    circle((R*np.cos(2*n*np.pi/m), R*np.sin(2*n*np.pi/m)), r, color[n])
plt.axis('equal')
plt.axis('off')
plt.show()
```
