# 花瓣生成的莲花
# lily Flower by Petals

```python
from n_lily import *
fig, ax = plt.subplots(figsize=(9, 9))
ax.set_aspect('equal')
plt.axis('off')

n = 4
N = 12
for i in range(0, N):
    one_petal((0, 0), 1, n, 2*i*np.pi/N, 'g')
    # one_petal_field((0, 0), 1, n, i*np.pi/6, 'y')
    one_petal((0, 0), np.sqrt(2*np.cos(np.pi/n))
              ** 2, n, 2*i*np.pi/N+np.pi/N, 'g')
    one_petal((0, 0), np.sqrt(2*np.cos(np.pi/4))
              ** 4, n, 2*i*np.pi/N, 'g')
```
