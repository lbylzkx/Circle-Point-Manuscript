```python
def symmetric_point(previous, point1, point2):
    """
    计算一点关于另外两点连线的对称点
    :param point1[0]: 第一个点的横坐标
    :param point1[1]: 第一个点的纵坐标
    :param point2[0]: 第二个点的横坐标
    :param point2[1]: 第二个点的纵坐标
    :param previous[0]: 待求对称点的横坐标
    :param previous[1]: 待求对称点的纵坐标
    :return: 对称点的横纵坐标
    """
    point_s = [0, 0]
    # 计算直线斜率和截距
    if (point2[0] - point1[0] != 0):
        k = (point2[1] - point1[1]) / (point2[0] - point1[0])
        b = point1[1] - k * point1[0]
        # 判断斜率是否为零
        if k == 0:
            # 两点连线水平，对称点横坐标不变，纵坐标为 2 * point1[1] - previous[1]
            point_s[0] = previous[0]
            point_s[1] = 2 * point1[1] - previous[1]
        else:
            # 计算对称点横纵坐标
            point_s[0] = 2 * (k * previous[1] - previous[0] + k * b) / \
                (k ** 2 + 1) - previous[0]
            point_s[1] = 2 * k * (k * previous[1] - previous[0] +
                                  b) / (k ** 2 + 1) - previous[1]
    else:
        # 两点连线垂直，对称点横坐标为 2 * point1[0] - previous[0]，纵坐标不变
        point_s[0] = 2 * point1[0] - previous[0]
        point_s[1] = previous[1]
    return point_s


# 示例
point1 = [-1, -1]
point2 = [0, 0]
previous = [2, 1]

point_s = symmetric_point(previous, point1, point2)
print(point_s)
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots(figsize=(9, 9))

R = 1
r = 2*np.sin(np.pi/12)

alpha = np.linspace(0, np.pi, 100)
x1 = np.cos(alpha)+R
y1 = np.sin(alpha)-r
plt.fill_between(x1, y1, where=(y1 >= 0), color='b')
beta = np.linspace(np.pi, 2*np.pi, 100)
x2 = np.cos(beta)+R
y2 = np.sin(beta)+r
plt.fill_between(x2, y2, where=(y2 <= 0), color='b', linewidth=2)

plt.plot(x1, y1, color='r')
plt.plot(x2, y2, color='r')
# 显示图表
plt.axis('equal')
plt.show()
```
